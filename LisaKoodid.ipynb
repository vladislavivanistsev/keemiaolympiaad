```python
from PyPDF2 import PdfFileWriter, PdfFileReader
from io import StringIO
import io
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
pdfmetrics.registerFont(TTFont('Corbel', 'Corbel.ttf'))
```

```python
def addcode(code,template,string):
    # read your existing PDF
    pdf = PdfFileReader(open(template, "rb"))
    npages = pdf.getNumPages()
    
    # creade the code stamp
    packet = io.BytesIO()
    can = canvas.Canvas(packet, pagesize=A4)
    for i in range(npages):
        can.setFont('Corbel', 12)
        can.drawString(420, 810, '{0}: {1}'.format(string,code))
        can.showPage()
    can.save()
    #move to the beginning of the StringIO buffer
    packet.seek(0)
    
    # create a new PDF with Reportlab
    new_pdf = PdfFileReader(packet)
    output = PdfFileWriter()
    # add the code the existing pages
    for i in range(npages):
        page = pdf.getPage(i)
        page.mergePage(new_pdf.getPage(i))
        output.addPage(page)
    # write "output" to file
    outputStream = open('{0}.pdf'.format(code), "wb")
    output.write(outputStream)
    outputStream.close()
```
